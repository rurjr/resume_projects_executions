```python
from enum import Enum
import sqlite3
# підключаєм базу данних
db = sqlite3.connect('home_task_4_sqlite3_movie_table.db')
# підключаєм курсор
cur = db.cursor()
# створуєм таблицю moveis
cur.execute("""CREATE TABLE IF NOT EXISTS MOVIES(
    title text,
    director text,
    genre text,
    release_date text,
    rating text
)""")   
db.commit()
 
 
# Enum для жарнів
class Genre(Enum):
    Action = 'Action'
    Drama = 'Drama'
    Adventure = 'Advanture'
    Fantacy = 'Fantacy'

# Cловник з усіма фільмами
movies = {'Title':['Avengers:Infinity War','The Flash','Spider-man: No Way Home'],'Director':[['Antony Russo','Joe Russo'],'Andy Muschietti','Jon Watts'] ,'Release_date':['27 April 2018','16 June 2023','17 December 2021'], 'Reception':[8.4,6.7,10.0],'Genre':[Genre.Action.value,Genre.Adventure.value,Genre.Fantacy.value]}
 
 
# Словник з усіма деректорами фільмім з нашого словника
directors = {'First_name':[['Antony ','Joe'],'Andy','Jon'],'Last_name':[['Russo','Russo'],'Muschietti','Watts'],'Age':[[54,54],50,42],'Short_info':[['Anthony Russo (born February 3, 1970) and Joseph Russo (born July 18, 1971), collectively known as the Russo brothers (/ˈruːsoʊ/ ROO-soh), are American directors, producers, and screenwriters. They direct most of their work together. They are best known for directing four films in the Marvel Cinematic Universe (MCU): Captain America: The Winter Soldier (2014), Captain America: Civil War (2016), Avengers: Infinity War (2018), and Avengers: Endgame (2019). Endgame grossed over $2.798 billion worldwide, briefly becoming the highest-grossing film of all time.','Anthony Russo (born February 3, 1970) and Joseph Russo (born July 18, 1971), collectively known as the Russo brothers (/ˈruːsoʊ/ ROO-soh), are American directors, producers, and screenwriters. They direct most of their work together. They are best known for directing four films in the Marvel Cinematic Universe (MCU): Captain America: The Winter Soldier (2014), Captain America: Civil War (2016), Avengers: Infinity War (2018), and Avengers: Endgame (2019). Endgame grossed over $2.798 billion worldwide, briefly becoming the highest-grossing film of all time.'],'В 2013 году американские киностудии Toma 78 и De Milo Productions пригласили Мускетти для полнометражной экранизации его же одноимённой короткометражки 2008 года «Мама». Главные роли в фильме исполнили Джессика Честейн и Николай Костер-Вальдау. При бюджете в 15 миллионов долларов[2][3] хоррор Энди собрал в мировом прокате свыше ста сорока пяти[3].','До художніх фільмів, займався зйомками реклами для компанії Park Pictures. Намагаючись отримати місце режисера фільму «Людина-павук: Повернення додому», Уоттс зробив татуювання головного героя на грудях для того, щоб «виділити [себе] серед інших [претиндентів]»[4].']}
 
 
# Клас для обєктів директор
class Director():

    def __init__(self,first_name:str,last_name,age:str,short_info:str) -> None:
        self.first_name = first_name
        self.last_name = last_name
        self.age = age
        self.short_info = short_info
    # метод для красивого зображення інформації про директора  
    def pretty_print_director_info(self):
        return  f' \n First_name: {self.first_name} \n Last_name: {self.last_name} \n Age: {self.age} \n Wikipedia : {self.short_info} \n'
 
    
# список з всіх дерескторів у формі обєкту класу Director     
list_of_directors_objects = [Director(directors['First_name'][i],directors['Last_name'][i],directors['Age'][i],directors['Short_info'][i]) for i in range(len(directors['First_name']))]
 
   
#клас movie
class movie():
    def __init__(self,title:str,release_date:int,director:list,reception:float,genre:str) -> None:
        self.title = title
        self.release_date = release_date
        self.director = director
        self.reception = reception
        self.genre = genre
        
        
     # метод для красивого зображення інформації про фільм       
    def pretty_print_movie_info(self):
        return f' \n Movie: {(self.title)} \n Release Date: {(self.release_date)} \n Director: {(self.director.first_name)}  {self.director.last_name} \n Reception: {(self.reception)}  \n Genre: {(self.genre)} \n '
    
    
    # метод для відображення усіх елементів бази данних
    def show():
        cur.execute("SELECT * FROM MOVIES")
        print(cur.fetchall())
        
         
    # метод для внесення данних у базу данних
    def Inserting(self):
        cur.execute("""INSERT INTO MOVIES VALUES(?,?,?,?,?)""",(self.title,str(self.director),str(self.genre),str(self.release_date),str(self.reception)))
        db.commit()
        
          
    # метод для пошуку конкретного фільму у базі данних
    def Selecting(title,variable,value):
        selecting_statement = f"SELECT * FROM MOVIES WHERE title = '{title}' AND {variable} = '{value}'"
        cur.execute(selecting_statement)
        print(cur.fetchall())
        
        
    # метод для видалення конекретного фільму з бази данних  
    def Deleting(title,variable,value):
        deleting_statement = f"DELETE FROM MOVIES WHERE title = '{title}' AND {variable} = '{value}'"
        cur.execute(deleting_statement)
        db.commit()
        
        
    # метод для змінни данних конеретного фільму який знаходиться у базі данних  
    def Updating(title,variable,innitial_value,new_value):
        updating_statement = f"UPDATE MOVIES SET {variable} = '{new_value}' WHERE title = '{title}' AND {variable} = '{innitial_value}'"
        cur.execute(updating_statement)
        db.commit()
        
        
    # метод для очищення бази данних  
    def clear():
        cur.execute("""DELETE FROM MOVIES""")
        db.commit()
        
          
# список з всіх фільмів у формі обєкту класу movie
list_of_movies_objects = [movie(movies['Title'][i],movies['Release_date'][i],list_of_directors_objects[i],movies['Reception'][i],movies['Genre'][i]) for i in range(len(movies['Title']))]
 

print(f'Для демонстрації спершу очистимо нашу базу данних') 
movie.clear()


print(f'Продемонструємо що метод clear дійсно очистив базу і там немає елементів')
movie.show()


print(f'\nДодамо усі фільми у базу данних і продемонструємо оновлену базу данних')
for i in range(len(list_of_movies_objects)):
    movie.Inserting(list_of_movies_objects[i])
movie.show()


print(f"\nЗнайдемо фільм який для прикладу називається 'The Flash' і є у базі данних(припустимо що ми не шукаємо фільмів які не є у базі данних)")
movie.Selecting('The Flash','title','The Flash')


print(f"\nПрипустимо нам потрібно видалити з бази данних фільм з назвою 'Spider-man: No Way Home', що і зробимо і продемонструємо оновлену базу данних,що фільм був дійсно видалений")
movie.Deleting('Spider-man: No Way Home','title','Spider-man: No Way Home')
movie.show()


print(f"\nПрипустимо нам потрібно фільму 'Avengers:Infinity War' замінити значиння рейтингу на слово 'Brilliant',що і зробимо і покажемо що значення рейтингу для цього фільму було змінено\n")
movie.Updating('Avengers:Infinity War','rating',8.4,'Brilliant')


print('Avengers:Infinity War')
movie.Selecting('Avengers:Infinity War','title','Avengers:Infinity War')


print(f'\nІ зауважимо що рейтинг інших фільмів не було змінено')
movie.show()


#P.S Для чого використовувалися змінні title,variable,value,innitial_value,new_value при створенні методів класу для роботи з таблицями?
# Все банально просто, і базується на узагальнювальній здатності нашого алгоритму, для того щоб не писати n окреми методів для знаходження,видалення... потрібного елементу
# по назві фільму, по директору і так далі, ми використувуєму універсальну змінну variable яка приймає всі потрібні значення змінних загальних назв і вже виконує фільмтрування безпосерденьо по ньому
# таким чином покриваючи усі можливі змінні назв  по якиз може проводитися запит + при змінні кількості змінних алгоритм продовжить працювати
# аналогічну концепцію мають за собою змінні value,innitial_value,new_value
# Також зауважу,що інколи може виникати на перший погляд дублювання коду як ось наприклад movie.Deleting('Spider-man: No Way Home','title','Spider-man: No Way Home')
# але ні, адже перше входження 'Spider-man: No Way Home' є статичним і певною мірою обовязковим, для вказування на множину фільмів з певною назвою,
# а друге входження 'Spider-man: No Way Home' зележить від вказання другого аргументу, і є опціональним (при другому параметрі title є при будь якому іншому немає) і використовується вже для конкретизації елементу
# з уже підмножини 'Spider-man: No Way Home'  множини усіх фільмів, адже фільми можуть мати однакову назву а бути зовсім різними,і тому ми не дублюємо код а конкретизуємо елемент який хочемо отримати


db.close()
```
